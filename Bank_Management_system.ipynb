```python
accounts = {
    "name": "sanaullah",
    "balance": 0.0,
    "transactions": []
}

counter = 0  # Keep track of the number of accounts processed

def deposit():
    global counter
    counter += 1

    # Input name
    name = input("Enter your name: ")

    # Input amount with validation for non-negative values
    while True:
        try:
            amount = float(input("Enter the amount: "))
            if amount <= 0:
                print("Amount must be greater than zero. Please try again.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    # Update account details
    accounts["name"] = name
    accounts["balance"] += amount
    accounts["transactions"].append(amount)
    current_balance = accounts["balance"]

    # Save account details to file
    with open("accounts.txt", "a") as f:
        f.write(f"""
  /////////////////////////// start Account {counter} /////////////////////////////////
  Name: {accounts["name"]}
  Balance: {accounts["balance"]}
  Transactions: {accounts["transactions"]}
  Current Balance: {current_balance}
  /////////////////////////// Next Account {counter} /////////////////////////////////
  """)

    return f"""
  Name: {accounts["name"]}
  Balance: {accounts["balance"]}
  Transactions: {accounts["transactions"]}
  """

# Try to execute the deposit function
try:
    print(deposit())
except Exception as e:
    print(f"An error occurred: {e}")
```

Step 3: Withdraw Money
The withdraw function allows the user to withdraw money from the account.
Function requirements:
 The withdrawal amount must be positive.
 The user must have enough balance to withdraw the money.
 If the withdrawal is successful, subtract the withdrawal amount from the balance and
record the transaction.
 If the balance is insufficient, print an error message.


```python
# Assuming accounts is a dictionary with 'name', 'balance', and 'transactions'.
accounts = {
    "name": "sanaullah",
    "balance": 1000.0,
    "transactions": []
}

def withdraw():
    try:
        # Taking user input for the withdrawal process
        amount = float(input("Enter the amount to withdraw: "))
        name = input("Enter your name: ").strip()  # Remove extra spaces from name input

        # Check if the amount is positive and if the name matches the account's name
        if amount > 0:
            if name == accounts["name"]:
                if amount <= accounts["balance"]:
                    # Update the balance and record the transaction
                    accounts["balance"] -= amount
                    accounts["transactions"].append(-amount)

                    # Write the updated account details to a file
                    with open("accounts.txt", "a") as f:
                        f.write(f"""
/////////////////////////// start Account /////////////////////////////////
Name: {accounts["name"]}
Balance: {accounts["balance"]}
Transactions: {accounts["transactions"]}
Current Balance: {accounts["balance"]}
/////////////////////////// End of Account /////////////////////////////////
""")

                    return f"Withdrawal successful!\nName: {accounts['name']}\nBalance: {accounts['balance']}\nTransactions: {accounts['transactions']}"
                else:
                    return "Insufficient balance"
            else:
                return "Account name does not match"
        else:
            return "Invalid amount entered. Please enter a positive number."

    except Exception as e:
        return f"An error occurred: {e}"

# Call the withdraw function and print the result
withdrawn_amount = withdraw()
print(withdrawn_amount)
```
